```python

# Dataset loader datastructure
class SolarDataset(Dataset):
    def __init__(self, df, root, transform=None, mode="sequence"):
        self.df = df
        self.root = root
        self.transform = transform
        self.mode = mode

        self.valid_indices = []

        # for all the data
        for i in range(len(df)):

            # get the row and sample id
            row = df.iloc[i]
            sample_id = row.name

            try:
                # get region and sample folders from df, 
                region, sample_folder = sample_id.split("_", 1)
                sample_path = os.path.join(root, region, sample_folder)

                # get images
                files = os.listdir(sample_path)

                # get magnetrogram images only for the time being
                mags = [f for f in files if "magnetogram" in f]

                # keep only 4-frame samples
                if len(mags) == 4:
                    self.valid_indices.append(i)

            except:
                continue

    def __len__(self):
        return len(self.valid_indices)

    # return images, 
    def __getitem__(self, idx):
        row = self.df.iloc[self.valid_indices[idx]]
        sample_id = row.name
        target = row["peak_flux"]

        region, sample_folder = sample_id.split("_", 1)
        sample_path = os.path.join(self.root, region, sample_folder)

        files = sorted(os.listdir(sample_path))
        mags = [f for f in files if "magnetogram" in f]

        imgs = []
        for f in mags:
            img = Image.open(os.path.join(sample_path, f)).convert("L")

            if self.transform:
                img = self.transform(img)

            imgs.append(img)
        if self.mode == "single":
            img = imgs[-1]  # baseline
            return img, torch.tensor(target, dtype=torch.float32)
    
        elif self.mode == "sequence":
            imgs = torch.stack(imgs)
            return imgs, torch.tensor(target, dtype=torch.float32)
    
        else:
            raise ValueError("mode must be 'single' or 'sequence'")
```

```python
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.nn.functional as F

class BaselineCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)

        self.pool = nn.MaxPool2d(2)

        self.adapt = nn.AdaptiveAvgPool2d((8, 8))

        self.fc1 = nn.Linear(32 * 8 * 8, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # 128 → 64
        x = self.pool(F.relu(self.conv2(x)))  # 64 → 32

        x = self.adapt(x)
        
        x = x.view(x.size(0), -1)

        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x.squeeze()
```

```python
train_loader = DataLoader(train_single, batch_size=16, shuffle=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = BaselineCNN().to(device)

loss_fn = nn.L1Loss()  # MAE (important for your dataset)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
epochs = 10

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for x, y in train_loader:
        x = x.to(device).float()
        y = y.to(device).float()

        optimizer.zero_grad()

        preds = model(x)
        loss = loss_fn(preds, y)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(train_loader):.6f}")
```

```python
from torch.utils.data import random_split

train_size = int(0.8 * len(train_single))
val_size = len(train_single) - train_size

train_ds, val_ds = random_split(train_single, [train_size, val_size])

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=16, shuffle=False)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = BaselineCNN().to(device)
loss_fn = nn.L1Loss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

epochs = 10


for epoch in range(epochs):
    
    # -------------------
    # TRAINING
    # -------------------
    model.train()
    train_loss = 0

    for x, y in train_loader:
        x = x.to(device).float()
        y = y.to(device).float()

        optimizer.zero_grad()

        preds = model(x)
        loss = loss_fn(preds, y)

        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(train_loader)

    # -------------------
    # VALIDATION
    # -------------------
    model.eval()
    val_loss = 0

    with torch.no_grad():
        for x, y in val_loader:
            x = x.to(device).float()
            y = y.to(device).float()

            preds = model(x)
            loss = loss_fn(preds, y)

            val_loss += loss.item()

    val_loss /= len(val_loader)

    # -------------------
    # PRINT
    # -------------------
    print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss:.6f} | Val Loss: {val_loss:.6f}")
```

```python
class FrameCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2)
        self.adapt = nn.AdaptiveAvgPool2d((8, 8))

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.adapt(x)
        x = x.view(x.size(0), -1)
        return x   # [B, 2048]
```

```python
class TemporalAttention(nn.Module):
    def __init__(self, feature_dim):
        super().__init__()
        self.attn = nn.Linear(feature_dim, 1)
    def forward(self, x):
        # x: [B, T, F]
        weights = self.attn(x)          # [B, T, 1]
        weights = torch.softmax(weights, dim=1)
        out = (x * weights).sum(dim=1)  # weighted sum
        return out, weights
```

```python
class AttentionTemporalCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.encoder = FrameCNN()
        self.attention = TemporalAttention(32 * 8 * 8)

        self.head = nn.Sequential(
            nn.Linear(32 * 8 * 8, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        # x: [B, T, C, H, W]
        B, T, C, H, W = x.shape

        features = []

        for t in range(T):
            f = self.encoder(x[:, t])
            features.append(f)
        features = torch.stack(features, dim=1)  # [B, T, F]
        fused, attn_weights = self.attention(features)
        out = self.head(fused).squeeze()

        return out, attn_weights
```
